```python
dtrain.sort_values(by=['price'], inplace=True, ascending = False)
dtrain
```

```python
dtrain = dtrain.drop(['city'], axis =1)
dtrain
```

```python
dtrain = dtrain.drop(dtrain.index[np.where(dtrain['carat'] > 0.9)])
dtrain = dtrain.drop(dtrain.index[np.where(dtrain['price'] < 500)])
dtrain = dtrain.drop(dtrain.index[np.where(dtrain['x'] == 0)])
dtrain = dtrain.drop(dtrain.index[np.where(dtrain['y'] == 0)])
dtrain = dtrain.drop(dtrain.index[np.where(dtrain['z'] == 0)])
dtrain = dtrain.drop(['index_id'], axis = 1)

dtrain
```

```python
cat_features = ['cut','color','clarity']
dtrain_one_hot_encoding = pd.get_dummies(dtrain[cat_features], 
                                          columns=['cut','color','clarity'], 
                                          drop_first=True)
dtrain_one_hot_encoding

target = dtrain['price']
```

```python
dtrain_one_hot_encoding.columns
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(dtrain_one_hot_encoding)
scaled_data
dtrain_one_hot_encoding
```

```python
target
```

```python
X_train, X_validation, y_train, y_validation = train_test_split(scaled_data, target, random_state = 42, test_size = 0.2)
```

```python
# Se entrena el modelo:

model = RandomForestRegressor()
model.fit(X = X_train, y = y_train)
```

```python
# MÃ©trica de error:

# Primero se hace predicciÃ³n del 20%:
y_pred = model.predict(X_validation)

mean_squared_error(y_validation, y_pred, squared=False)
```
